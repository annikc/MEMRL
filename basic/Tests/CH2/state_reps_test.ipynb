```python
import gym
import sys
import numpy as np
import matplotlib.pyplot as plt
sys.path.append('../../modules/')
from modules.Agents.RepresentationLearning.learned_representations import place_cell, rand_place_cell, sr, onehot
```

```python
env = gym.make('gridworld:gridworld-v1')
index = env.state
print(index)
```

```python
### onehots
oh_reps, name, dim, _ = onehot(env)
# get "centres" a la place cells 
i = np.arange(0,1,0.05)
j = np.arange(0,1,0.05)
centres = []
for ii in i:
    for jj in j:
        centres.append([ii,jj])
centres = np.asarray(centres)

OH = np.zeros((env.nstates, env.nstates))
for state in oh_reps.keys():
    OH[state] = oh_reps[state]

# get info for single state 
rep = oh_reps[index]

# plot state representation vector
plt.figure(figsize=(6,0.75))
plt.imshow([rep], aspect='auto', interpolation='none')
plt.xticks(rotation=90)
plt.gca().get_yaxis().set_visible(False)
plt.show()

# plot reshaped state representation
plt.figure()
ax = plt.imshow(rep.reshape(*env.shape))
plt.xticks(np.arange(0,20,5),labels=[0,5,10,15])
plt.yticks(np.arange(0,20,5),labels=[0,5,10,15])
plt.colorbar(ax)

# try centres
plt.figure(figsize=(3.5,3.5))
plt.scatter(centres[:,1]+0.025, centres[:,0]+0.025,c=rep, s=120)
plt.axis('equal')
plt.gca().invert_yaxis()
plt.tight_layout()

# get full matrix 
plt.figure()
ax = plt.imshow(OH)
plt.colorbar(ax)

plt.show()
```

```python
### state-centred place cells 
pc_reps, name, dim, centres = place_cell(env, field_size=0.05)

PC = np.zeros((env.nstates, env.nstates))
for state in pc_reps.keys():
    PC[state] = pc_reps[state]

# get info for single state 
rep = pc_reps[index]

# plot state representation vector
plt.figure(figsize=(6,0.75))
ax = plt.imshow([rep], aspect='auto',interpolation='none')
plt.xticks(rotation=90)
plt.gca().get_yaxis().set_visible(False)

# plot reshaped state representation
plt.figure()
ax = plt.imshow(rep.reshape(*env.shape))
plt.colorbar(ax)
plt.xticks(np.arange(0,20,5),labels=[0,5,10,15])
plt.yticks(np.arange(0,20,5),labels=[0,5,10,15])
plt.show()

# try centres
plt.figure(figsize=(3.5,3.5))
plt.scatter(centres[:,1]+0.025, centres[:,0]+0.025,c=rep, s=120, marker='o')
plt.axis('equal')
plt.gca().invert_yaxis()
plt.tight_layout()


# get full matrix 
plt.figure()
plt.imshow(PC[0:100,0:100])
plt.show()
```
